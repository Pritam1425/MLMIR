```
import librosa as lr
from glob import glob
```

```
#you need to replace 2 by your own folder which are listed in above block
audio_files = glob('2' + '/*.mp3')
```

```
#if len comes as 0 then don't go further there's a problem
len(audio_files)
```

```
os.chdir('/content/gdrive/My Drive/MIR')
```

```
data_array = []
```

```
os.chdir('/content/gdrive/My Drive/MIR/magnatagatune')
for i in range (len(audio_files)):
  y, sr = lr.load(audio_files[i])
  arr = lr.feature.melspectrogram(y=y, sr=sr)
  arr = arr.reshape(1,128,1255)
  #print(arr.shape)
  data_array.append(arr)
  if i%20==0:
    print('Processed: '+str(i))
```

```
data = np.array(data_array)
```
